#### 遍历logs/train下面的图片

```python
import os
from PIL import Image
```

```python
folder_path = "/research/d5/gds/xywen22/project/HumanSDs-MAE-map-conv-newdata-all/logs/2023-08-23T15-09-25_humansd-finetune-originalloss/images/train"
save_path = "/research/d5/gds/xywen22/project/HumanSDs-MAE-map-conv-newdata-all/logs/2023-08-23T15-09-25_humansd-finetune-originalloss/merge"
```

```python
for i in range(4800, 6000, 10):
    print(i)
    for file_name in os.listdir(folder_path):
        if "conditioning_gs-{:06d}".format(i) in file_name:
            # print(file_name)
            # conditioning_gs-005700_e-000950_b-000000.png
            img_condition = Image.open(os.path.join(folder_path, file_name))
            suffix = file_name.split("conditioning_gs-")[1]
            img_input = "inputs_gs-" + suffix
            img_sample_cfg = "pose_samples_cfg_scale_5.00_gs-" + suffix
            img_sample = "pose_samples_gs-" + suffix

            img_input = Image.open(os.path.join(folder_path, img_input))
            img_sample_cfg = Image.open(os.path.join(folder_path, img_sample_cfg))
            img_sample = Image.open(os.path.join(folder_path, img_sample))
            # img.show()

            # merge to one image
            width, height = img_condition.size
            new_im = Image.new('RGB', (width*2, height*2))
            new_im.paste(img_condition, (0, 0))
            new_im.paste(img_input, (width, 0))
            new_im.paste(img_sample_cfg, (0, height))
            new_im.paste(img_sample, (width, height))
            new_im.save(os.path.join(save_path, "merge_" + file_name))

            break
```
